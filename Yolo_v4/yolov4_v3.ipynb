```python
def new_cols(df):
  df['classNum'] = ''
  df["x"] = ''
  df["y"] = ''
  df['w'] = ''
  df['h'] = ''
  return df
```

```python
new_cols(train_yolo)
```

```python
new_cols(test_yolo)
```

```python
classes = {'ESP8266': 0, 'Heltec_ESP32_Lora': 1, 'Raspberry_Pi_3': 2, 'Arduino_Nano': 3}
```

```python
def class_num(df):
  for i in range(len(df['filename'])):
    df['classNum'] = df["class"].map(classes)
  return df
```

```python
class_num(train_yolo)
```

```python
class_num(test_yolo)
```

```python
def count_bbox(df):
  for i in range(len(df['filename'])):
    train_yolo['x'] = (((df["xmin"] + df["xmax"]) /2) * (1/df["width"])).round(3).astype(float)
    df['y'] = (((df["ymin"] + df["ymax"]) /2) * (1/df["height"])).round(3).astype(float)
    df['w'] = ((df["xmax"] - df["xmin"]) * (1/ df["width"])).round(3).astype(float)
    df['h'] = ((df["ymax"] - df["ymin"]) * (1/ df["height"])).round(3).astype(float)
  return df
```

```python
count_bbox(train_yolo)
```

```python
count_bbox(test_yolo)
```

```python
# Create .txt files 
def create_txt(img_path, df):
  os.chdir(img_path)

  for current_dir, dirs, files in os.walk('.'):
    for f in files:
      file_name = f
      yolo_file = df.loc[df['filename'] == file_name]
      txt_files = yolo_file.loc[:, ['classNum', 'x', 'y', 'w', 'h']].copy()
      save_path = img_path + "/" + file_name[:-4] + '.txt'
      txt_files.to_csv(save_path, header=False, index=False, sep=' ')
```
